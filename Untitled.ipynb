# Flop Counting & Performance Plot for Basic Implementation
<hr>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sure(n):
    adds = 12*n**2;
    mults = 12*n**2
    divs = 0
    return np.array([adds, mults, divs])

def flt_channel_basic(f, r, n):
    adds = (n-2*(r+f))**2 * (2*r+1)**2 * (18*(2*f+1)**2 + 4) + (n-2*(r+f))**2
    mults = (n-2*(r+f))**2 * (2*r+1)**2 * (9*(2*f+1)**2 + 5) 
    divs = (n-2*(r+f))**2 * (2*r+1)**2 * (3*(2*f+1)**2+1) + (n-2*(r+f))**2 
    return np.array([adds, mults, divs])


def flt_buffer_basic(f, r, n):
    adds = (n-2*(r+f))**2 * (2*r+1)**2 * (18*(2*f+1)**2 + 6) + 3*(n-2*(r+f))**2
    mults = (n-2*(r+f))**2 * (2*r+1)**2 * (9*(2*f+1)**2 + 7) 
    divs = (n-2*(r+f))**2 * (2*r+1)**2 * (3*(2*f+1)**2+1) + 3*(n-2*(r+f))**2
    return np.array([adds, mults, divs])

def flt(f,r,n):
    adds = (n-2*(r+f))**2 * (2*r+1)**2 * (18*(2*f+1)**2 + 15) + 8*(n-2*(r+f))**2 
    mults = (n-2*(r+f))**2 * (2*r+1)**2 * (9*(2*f+1)**2 + 16) + 12*(n-2*(r+f))**2
    divs = (n-2*(r+f))**2 * (2*r+1)**2 * (3*(2*f+1)**2+4) + 3*(n-2*(r+f))**2 
    return np.array([adds, mults, divs])

def filter_averaging(n):
    adds = 12*n**2
    mults = 9*n**2
    divs = 9*n**2
    return np.array([adds, mults, divs])
```

```python
def basic(n, R):

    # Compute op-count for one run of the basic implementation
    total = 2*flt_buffer_basic(3, 5, n)
    total += flt(1, R, n)
    total += flt(3, R, n)
    total += flt(1, R, n)
    total += (3*sure(n))
    total += (3*flt_channel_basic(1,1,1))
    # total += binary_selection(n)
    total += flt_channel_basic(1, 5, n)
    total += filter_averaging(n)
    
    return total
```

```python
## Store measured cycles
n = np.array([32, 64, 80, 96, 112, 128, 256, 512, 1024])
cycles_O0 = [8.09516 * 10**8,
             1.14798 * 10**10,
             2.07577 * 10**10,
             3.45336 * 10**10,
             4.98050 * 10**10,
             7.37735 * 10**10, 
             3.45199 * 10**11, 
             1.49453 * 10**12, 
             6.31558 * 10**12]

cycles_O3_noVectorize = [1.19545 * 10**8,
                         1.60791 * 10**9,
                         2.91836 * 10**9,
                         4.71216 * 10**9,
                         7.03140 * 10**9,
                         1.07667 * 10**10, 
                         5.77562 * 10**10, 
                         2.45166 * 10**11, 
                         1.08805 * 10**12]

cycles_O3_all = [1.21228 * 10**8,
                 1.58903 * 10**9,
                 3.13656 * 10**9,
                 5.07768 * 10**9,
                 7.26149 * 10**9,
                 1.04548 * 10**10,
                 5.61946 * 10**10, 
                 2.50940 * 10**11, 
                 1.11169 * 10**12]
```

```python
## Compute Performance
R = 10
performance_O0 = []
performance_O3_noVectorize = []
performance_O3_all = []

for i in range(n.shape[0]):
    flop_count = basic(n[i],R).sum()
    performance_O0.append(flop_count / cycles_O0[i])
    performance_O3_noVectorize.append(flop_count / cycles_O3_noVectorize[i])
    performance_O3_all.append(flop_count / cycles_O3_all[i])
```

```python
# ######################################################
# PERFORMANCE PLOT
# => Basic Implementation with different compiler flags
# #####################################################

plt.figure(figsize=(20,10))

# Plot series
plt.plot(n, performance_O0, 'x--', color="C0", label="-O0")
plt.plot(n, performance_O3_noVectorize, 'x--', color="C2", label="-03 -fno-tree-vectorize")
plt.plot(n, performance_O3_all, 'x--', color="C1", label="-03 -ffast-math -march=native")

plt.xticks(n)
plt.ylim(0,2)
plt.xlim(20,1044)
plt.ylabel("Performance [Flops/Cycle]")
plt.xlabel("Input Size n: (n x n image)")

plt.suptitle("Performance Plot (R=10)", x=0.125, y=1.03, fontsize=20, ha='left')
plt.title("Intel(R) Core(TM) i9-8950HK CPU @ 2.90GHz \n" \
          "Cache Sizes - L1: 32KB, L2: 256KB, L3: 12MB \n" \
          "Compiler: Apple clang version 11.0.0 \n" \
          "OS: MacOS Catalina, 10.15.4", fontsize=12, x=0, y=1.04, ha='left') 

plt.legend()
plt.show()
```

```python
### Varying R with fixed n=128
R_array = np.array([3,5,7,9,11,13,15])
cycles_128 = [5.50518 * 10**9,
              7.23124 * 10**9,
              9.00504 * 10**9,
              1.17229 * 10**10,
              1.40053 * 10**10,
              1.63303 * 10**10,
              1.84790 * 10**10] 

cycles_64 =  [9.73642 * 10**8,
              1.21817 * 10**9,
              1.54581 * 10**9,
              1.72570 * 10**9,
              1.99203 * 10**9,
              2.12531 * 10**9,
              2.09615 * 10**9] 

cycles_32 =  [1.28231 * 10**8,
              1.46330 * 10**8,
              1.52661 * 10**8,
              1.44137 * 10**8,
              1.18311 * 10**8,
              9.95811 * 10**7,
              9.28873 * 10**7] 
         
```

```python
## Compute Performance
performance_128 = []
performance_64 = []
performance_32 = []

for i in range(R_array.shape[0]):
    flop_count_128 = basic(128,R_array[i]).sum()
    flop_count_64 = basic(64,R_array[i]).sum()
    flop_count_32 = basic(32,R_array[i]).sum()
    performance_128.append(flop_count_128 / cycles_128[i])
    performance_64.append(flop_count_64 / cycles_64[i])
    performance_32.append(flop_count_32 / cycles_32[i])
```

```python
plt.figure(figsize=(10,10))

# Plot series
plt.plot(R_array, performance_128, 'x--', color="C0", label="n=128")
plt.plot(R_array, performance_64, 'x--', color="C1", label="n=64")
plt.plot(R_array, performance_32, 'x--', color="C2", label="n=32")

plt.xticks(R_array)
plt.ylim(0,4)
plt.ylabel("Performance [Flops/Cycle]")
plt.xlabel("Window Radius R:")

plt.suptitle("Performance Plot", x=0.125, y=1.03, fontsize=20, ha='left')
plt.title("Intel(R) Core(TM) i9-8950HK CPU @ 2.90GHz \n" \
          "Cache Sizes - L1: 32KB, L2: 256KB, L3: 12MB \n" \
          "Compiler: Apple clang version 11.0.0 \n" \
          "OS: MacOS Catalina, 10.15.4", fontsize=12, x=0, y=1.04, ha='left') 

plt.legend()
plt.show()
```
